```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the data frame
data = pd.read_csv('Dataset.csv')

# Preprocess the data
data['Combined'] = data['Age'].astype(str) + ' ' + data['Genre'] + ' ' + data['Cast']

# Create the vectorizer and compute the feature matrix
vectorizer = CountVectorizer()
combined_matrix = vectorizer.fit_transform(data['Combined'])

# Function to recommend movies based on age, genre, and cast
def recommend_movies(age, genre, cast, num_recommendations=5):
    input_features = str(age) + ' ' + genre + ' ' + cast

    # Compute the cosine similarity between the input features and all movies
    input_vector = vectorizer.transform([input_features])
    cosine_sim = cosine_similarity(input_vector, combined_matrix)

    # Get the indices of movies with matching cast
    cast_indices = data[data['Cast'].str.contains(cast)].index

    # Filter the cosine similarity matrix for movies with matching cast
    filtered_cosine_sim = cosine_sim[:, cast_indices]

    # Get the top recommendations
    top_indices = filtered_cosine_sim.argsort()[0][::-1][:num_recommendations]
    recommendations = data.iloc[cast_indices[top_indices]]['Name']

    return recommendations

# Example usage
selected_age = input("Enter your age: ")
selected_genre = input("Enter your preferred genre: ")
selected_cast = input("Enter your preferred cast member: ")

recommendations = recommend_movies(selected_age, selected_genre, selected_cast)
print("Recommended movies:")
for movie in recommendations:
    print(movie)
```
